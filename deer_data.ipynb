```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize=(10, 8))
```

```python
 # True function, w/ some noise
def f(x):
    return 1.8 * x + np.random.normal(0, 3, x.shape[0]) + 50

xs = np.linspace(0,100,100)
ys = f(xs)
plt.scatter(x=xs,
            y=ys,
            color='green')
plt.xlabel('deer weight (kg)')
plt.ylabel('deer height (cm)')
plt.ylim(0, 300)
```
